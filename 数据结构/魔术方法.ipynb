```python
# 方法2: 使用第三方库 blinker（推荐用于生产环境）

# 首先需要安装: pip install blinker

try:
    from blinker import signal
    
    # 创建信号
    login_signal = signal('user-login')
    logout_signal = signal('user-logout')
    
    # 注册监听器（使用装饰器）
    @login_signal.connect
    def handle_login(sender, username, **kwargs):
        print(f"[blinker] 用户 {username} 登录了，发送者: {sender}")
    
    # 注册监听器（使用 connect 方法）
    def handle_logout(sender, username):
        print(f"[blinker] 用户 {username} 登出了")
    
    logout_signal.connect(handle_logout)
    
    # 发送信号
    print("=== blinker 信号示例 ===")
    login_signal.send('system', username='张三')
    logout_signal.send('system', username='李四')
    
    # 临时连接（只执行一次）
    @login_signal.connect_via('temp')
    def temp_handler(sender, username, **kwargs):
        print(f"[blinker] 临时处理: {username}")
    
    login_signal.send('temp', username='临时用户')
    login_signal.send('temp', username='临时用户2')  # 不会执行
    
except ImportError:
    print("blinker 未安装，请运行: pip install blinker")
    print("blinker 是一个强大的信号库，支持命名空间、弱引用等功能")
```
