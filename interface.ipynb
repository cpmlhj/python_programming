```python
# 单例
class SimpleClass:
    _variable_value = None

    def __init__(self, var) -> None:
        self.get_variable = var

    @property
    def get_variable(self):
        # 如果要写的是单例属性， 要显示为 self.__class__._variable_value
        return self.__class__._variable_value

    @get_variable.setter
    def get_variable(self, var):
        self.__class__._variable_value = var

    def display(self):
        return self.get_variable    


c1 = SimpleClass(1)
c2 = SimpleClass(2)

c2.get_variable = 5


print(SimpleClass.__dict__)
print(c1.__dict__)
```
