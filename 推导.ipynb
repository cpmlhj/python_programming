```python
# 列表推导 1  生成一个列表， 包含1~20 中所有能被3整除的数

nums = [
    n for n in range(1,20) if n % 3 == 0
]

print(nums)
```

```python
# 列表推导2 
"""
给定一个字符串列表  words = ['hello', 'world', 'python'] ，生成一个新列表，包含每个单词首字母大写后的结果
"""
pre_list = ['hello', 'world', 'python']

next_list = [
    next.capitalize() for next in pre_list
]

print(next_list)
```

```python
# 字典推导 给定一个列表  nums = [1, 2, 3, 4, 5] ，生成一个字典，键是数字，值是该数字的立方

nums = [1,2,3,4,5]

dict_nums = {
    (x): x **3 for x in nums
}

print(dict_nums)
```

```python
# 字典推导，给定一个字符串  s = "abracadabra" ，生成一个字典，统计每个字母出现的次数

s = "abracadabra"

print(s.count("a"))

count_s = {
    (x) : s.count(x) for x in set(s)
}

print(count_s)
```

```python
# 字典推导 翻转字典

original = {'a': 1, 'b': 2, 'c': 3}

xa = {
   x[1]: x[0] for x in original.items()
}

print(xa)
```

```python

original_1 = {'a': 1, 'b': 2, 'c': 1}

group = {}

for k,v in original_1.items():
     group.setdefault(v, []).append(k)

print(group)     
```
