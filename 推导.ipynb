```python
# 列表推导 1  生成一个列表， 包含1~20 中所有能被3整除的数

nums = [
    n for n in range(1,20) if n % 3 == 0
]

print(nums)
```

```python
# 列表推导2 
"""
给定一个字符串列表  words = ['hello', 'world', 'python'] ，生成一个新列表，包含每个单词首字母大写后的结果
"""
pre_list = ['hello', 'world', 'python']

next_list = [
    next.capitalize() for next in pre_list
]

print(next_list)
```
