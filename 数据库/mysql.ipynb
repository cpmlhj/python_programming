```python

from pymysql import connect, cursors

conn = connect(
    host="localhost",
    port=3306, 
    user="root", 
    passwd="123456", 
    database="backend_jdy",
    cursorclass=cursors.DictCursor)

with conn.cursor() as db:
    # 读取
    sql = "SELECT `id`  FROM `User` WHERE `username`=%s"
    db.execute(sql, ('admin' ))
    result = db.fetchone()
    print(result)
```
